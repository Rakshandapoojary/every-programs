```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:/Users/Admin/Desktop/Cars93.csv')
df
```

```python
fig=plt.figure(figsize=[4,4])
ax=fig.add_axes([0,0,1,1])
ax.set_title('',fontsize=15)
ax.set_xlabel('values',fontsize=15)
ax.set_ylabel('needs',fontsize=15)
ax.scatter(df['Horsepower'],df['MPG.city'],color='C3')
fig.show()
```

```python
fig=plt.figure(figsize=[4,4])
ax=fig.add_axes([0,0,1,1])
ax.set_title('distribution',fontsize=10)
ax.set_xlabel('values',fontsize=15)
ax.set_ylabel('needs',fontsize=15)
ax.hist(df['MPG.city'],color='C4',bins=3)
fig.show()
```

```python
fig=plt.figure(figsize=[4,4])
ax=fig.add_axes([0,0,1,1])
ax.set_title('distribution',fontsize=15)
ax.set_xlabel('values',fontsize=15)
ax.set_ylabel('needs',fontsize=15)
ax.boxplot(df['Price'])
fig.show()
```
