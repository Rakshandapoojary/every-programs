```python
import pandas as pd
df1=pd.read_csv('C:/Users/Admin/Desktop/mtcars.csv')
df1
```

```python
df1.info()
```

```python
import matplotlib.pyplot as plt
plt.hist(df1['mpg'],bins=10,color="C3")
plt.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=[6,4])
ax=fig.add_axes([0,0,1,1])
ax.hist(df1['mpg'],bins=12)
ax.set_title('Distribution',fontsize=16,color="C3")
ax.set_ylabel('frequency',fontsize=10)
ax.set_xlabel('wt',fontsize=10)
fig.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=[6,4])
ax=fig.add_axes([0,0,1,1])
ax.scatter(df1['mpg'],df1['wt'])
ax.set_title('Distribution',fontsize=16,color="C3")
ax.set_ylabel('weight',fontsize=10)
ax.set_xlabel('qsec',fontsize=10)
fig.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=[6,4])
ax=fig.add_axes([0,0,1,1])
ax.bar(df1['mpg'],df1['wt'])
ax.set_title('Distribution',fontsize=16,color="C3")
ax.set_ylabel('frequecy',fontsize=10)
ax.set_xlabel('am',fontsize=10)
fig.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=[6,4])
ax=fig.add_axes([0,0,1,1])
ax.boxplot(df1['mpg'])
ax.set_title('Distribution',fontsize=16,color="C3")
ax.set_ylabel('lenght',fontsize=10)
ax.set_xlabel('weight',fontsize=10)
fig.show()
```
