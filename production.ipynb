```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C:/Users/Admin/Desktop/crop_production.csv')
data
```

```python
data.info()
```

```python
data['Production'].fillna(data['Production'].mean(),inplace=True)
data
```

```python
data.isnull().sum(0)
```

```python
data.describe()
```

```python
cat_feature=['State_Name','District_Name','Season','Crop']
data=pd.get_dummies(data,columns=cat_feature)
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(data[["Crop_Year","Area","Production"]])
data_normalization=scaler.transform(data[["Crop_Year","Area","Production"]])
data_normalization
```

```python
data_normalization=pd.DataFrame(data_normalization,columns=["Crop_Year","Area","Production"])
data_normalization=data_normalization.join(data[data.columns.drop(["Crop_Year","Area","Production"])])
data_normalization
```

```python
x=data_normalization[data_normalization.columns.drop("Production")]
y=data_normalization["Production"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
train_score=model.score(x_train,y_train)
test_score=model.score(x_train,y_train)
print("Train(RSquare):",train_score)
print("Test(RSquare):",test_score)
```

```python
from sklearn.metrics import mean_squared_error
train_prediction=model.predict(x_train)
train_RMSE=mean_squared_error(y_train,train_prediction)**0.5
test_prediction=model.predict(x_test)
test_RMSE=mean_squared_error(y_test,test_prediction)**0.5
print("Train:",train_RMSE)
print("test:",test_RMSE)
```
