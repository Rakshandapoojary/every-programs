```python
fig=plt.figure(figsize=[3,3])
ax=fig.add_axes([0,0,1,1])
ax.set_title('distribution')
ax.boxplot(data['mpg'])
ax.set_ylabel('frequency')
ax.set_xlabel('time')
fig.show()
```

```python
for i in data['mpg']:
    q1=data['mpg'].quantile(0.25)
    q3=data['mpg'].quantile(0.75)
    iqr=q3-q1
    lower_limit=q1-1.5*iqr
    upper_limit=q3+1.5*iqr
    if i>upper_limit or i<lower_limit:
     data['mpg']=data['mpg'].drop(i)
plt.boxplot(data['mpg'])
plt.show()
```

```python
import numpy as np
import pandas as pd
Series_A=pd.Series([1,2,3,4,5])
Series_B=pd.Series([4,5,6,7,8])
union=pd.Series(np.union1d(Series_A,Series_B))
print(union)
```

```python
intersect=pd.Series(np.intersect1d(Series_A,Series_B))
intersect
```

```python
commonSeries=union[union.isin(intersect)]
commonSeries
```
