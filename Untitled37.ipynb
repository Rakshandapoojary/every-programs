```python
from sklearn.preprocessing import StandardScaler
# finding the mean and std deviation of numerical columns
scaler = StandardScaler().fit(data[['FEB', 'MAR']]) 
# scaling columns to a common range
data_normalized = scaler.transform(data[['FEB','MAR']])# Scaled data 
#note that output is a 2-Dimensional array of normalized data
data_normalized
```

```python
#updating the numerical(normalized) data into the dataframe
data_normalized=pd.DataFrame(data_normalized,columns=['FEB', 'MAR'])
data_normalized=data_normalized.join(data[data.columns.drop(['FEB', 'MAR'])])
data_normalized
```
