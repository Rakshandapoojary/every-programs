```python
data['Mins'].fillna(data['Mins'].mean(),inplace=True)
data
```

```python
data.columns
```

```python
data.info()
```

```python
# encoding categorical values
cat_features = ["Dismissal"]
data = pd.get_dummies(data,columns=cat_features)
data
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(data[['Batsman','Runs','Mins','BF','4s','6s','SR','Pos','Inns']]) 

data_normalized = scaler.transform(data[['Batsman','Runs','Mins','BF','4s','6s','SR','Pos','Inns']])
```

```python
data_normalized=pd.DataFrame(data_normalized,columns=['Batsman','Runs','Mins','BF','4s','6s','SR','Pos','Inns'])
data_normalized=data_normalized.join(data[data.columns.drop(['Batsman','Runs','Mins','BF','4s','6s','SR','Pos','Inns'])])
data_normalized
```

```python
Y = data_normalized['Runs']
X = data_normalized[data_normalized.columns.drop('Runs')]
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=100)
X_train.shape , X_test.shape

```

```python
#importing required module to build the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
train_score=model.score(X_train,Y_train)
test_score=model.score(X_test,Y_test)
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

```python
from sklearn.metrics import mean_squared_error
#root mean square error (RMSE) calculation for train data
train_predictions = model.predict(X_train)
train_RMSE=mean_squared_error(Y_train,train_predictions)**0.5
#root mean square error (RMSE) calculation for test data
test_predictions = model.predict(X_test)
test_RMSE=mean_squared_error(Y_test,test_predictions)**0.5
print('Train RMSE : ',train_RMSE)
print('Test RMSE : ',test_RMSE)


```
